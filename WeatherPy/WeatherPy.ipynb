```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count 
print(len(cities))
```

```python
 # Create search query, make request and store in json
    
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Set empty lists to hold 'Temperature (F)', 'Humidity (%)', 'Cloudiness (%)', 'Wind Speed (mph)'
#                       , 'Latitude', 'Northern Hemisphere' or 'Southern Hemisphere'

city = []
con = []
cid =   []
Temp = []
maxt = []
Hum =  [] 
Cloud = []
Wind =  []
Lat =   []
Lon =   []
date =[]

#print(cities)

i = 1
s = 1

print('Beginning Data Retrieval')     
print('-----------------------------')

#Checking data
#first_json = True

for c in cities: 
    #print(c)
    query = f"{url}&units={units}&appid={weather_api_key}&q=" + c
    response = requests.get(query)
    response_json = response.json()
    
#Checking data_1
#    if first_json:
#        print(response_json)
#        first_json = False
    
    try:
        if response.status_code == 200:
            city.append(c)
            con.append(response_json["sys"]["country"])
            cid.append(response_json['id'])
            Temp.append(response_json["main"]["temp"])
            maxt.append(response_json["main"]["temp_max"])
            Hum.append(response_json['main']['humidity'])
            Cloud.append(response_json['clouds']['all'])
            Wind.append(response_json['wind']['speed'])
            Lat.append(response_json['coord']['lat'])
            Lon.append(response_json['coord']['lon'])
            date.append(response_json['dt'])
            print(f'Processing Record {i} of Set {s} | {c}')
            if i < 50:
                i += 1
            else:
                i = 1
                s += 1
        else:
            print("City no Found. Skipping...")
    except KeyError as e:
            
            continue        
#    except Exception as e2:
#        raise e2
print('-----------------------------')
print('Data Retrieval Complete')     
print('-----------------------------')


#print("cities " + str(len(cities)))
#print('city ' + str(len(city)))
#print('id ' + str(len(cid)))
#print("temp " + str(len(Temp)))
#print('max temp ' + str(len(maxt)))
#print('hum '+ str(len(Hum)))
#print('cloud ' + str(len(Cloud)))
#print('win '+ str(len(Wind)))
#print('lat '+ str(len(Lat)))
#print('lon '+ str(len(Lon)))
#print('date '+ str(len(date)))
```

```python
# Create DataFrame
city_df = pd.DataFrame({
    'City': city,
    'Latitude': Lat,
    'Longitude':Lon,
    'Max Tem' : maxt,
    'Temperature (F)': Temp,
    'Humidity (%)': Hum,
    'Cloudiness (%)': Cloud,
    'Wind Speed (mph)': Wind,
    'Country' : con,
    'ID': cid,
    'Date' :date
})

city_df.to_csv(output_data_file, index=False, header=True)
city_df
```

## Linear Regression

```python
#Define de Df based on 'Latitude'
northern = newcity_df.loc[newcity_df["Latitude"] >= 0.0]
northern.reset_index(inplace=True)



southern = newcity_df.loc[newcity_df["Latitude"] < 0.0]
southern.reset_index(inplace=True)

#ss = northern.loc[northern["Latitude"] < 0.0]
#ss
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Get the plot
ts = time.gmtime(newcity_df['Date'].max())
y1 = northern['Max Tem']
x1 = northern['Latitude']


datestr = time.strftime("%x",ts)
plt.scatter(x1,y1, marker="o", facecolors="green", edgecolors="black",s=30, alpha=0.75)
#plt.grid()
plt.title(f"City Latitude vs. Max Temperature of Northern Hemisphere ({datestr})")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")

#Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x1,y1)
print(f"The r-squared is: {round(rvalue**2,4)}")
regress_values = x1 * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.plot(x1,regress_values,"r-")
plt.annotate(line_eq,(20,-10),fontsize=15,color="red")
plt.savefig(f"output_data/fig{fignum}.png")
plt.show()   
fignum = fignum + 1 
```
